# Group by

```python
df = pd.read_csv('/content/drive/MyDrive/Data/census.csv')
df = df[df['SUMLEV']==50]
df
```

```python
df['STNAME']
```

```python
%%timeit -n 1

for state in df['STNAME'].unique():
    avg_1 = (sum(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
            /len(df.where(df['STNAME']==state).dropna()['CENSUS2010POP']))
    
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print('Counties in state ' + state + ' have an average population of ' + str(avg) +' '+ str(avg_1))
```

```python
df.groupby?
```

```python
%%timeit -n 1

for group, frame in df.groupby('STNAME'):
    #print(group)
    #print(frame)
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group + ' have an average population of ' + str(avg))
```

```python
df = df.set_index('STNAME')

def fun(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2

for group, frame in df.groupby(fun):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing.')
```
